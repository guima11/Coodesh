##4. Mês com mais adições na Netflix

```python
Meses = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
date = netflix['date_added'].to_string().split(' ')
r = 0
Y = np.zeros(12)
for mes in Meses:
    data = date.count(mes)
    Y[r] = data
    r += 1

Meses = pd.DataFrame(data = [Meses, Y])
Meses = Meses.T
Meses.columns = ["Mês", "Número de Adições"]
Meses
```

```python
Meses = Meses.sort_values(by = 'Número de Adições', ascending = False)
Meses.head()
```

Com isso, pode-se observar que o mês com amior entrada de filmes no Netflix é o mês de julho com 827, seguido de dezembro com 813 que são so meses considerados de férias escolares.

###7.1. e 8.1. Geral

Visando observar de forma mais acertiva tais resultados, tomou-se como foco o uso de gráfico de barras para comparativo entre ambas as plataformas.

```python
filme_geral = dados[fg]['type']
filme_geral = filme_geral.count()
filme_geral
```

```python
sg = dados['type'].isin(Show)
serie_geral = dados[sg]['type'].count()
serie_geral
```

```python
g = {
    'Entretenimento' : ['Movies', 'TV Shows'],
     'valor' : [filme_geral, serie_geral]}
g = pd.DataFrame(g)
g
```

```python
sns.barplot(x = "Entretenimento", y = "valor", data = g)
```

Observa-se que há uma publicação muito maior de filmes que de séries de uma forma geral nas plataformas de *"streamming"*

###7.2. e 8.2. Amazon Prime

```python
fa = amazon['type'].isin(Filme)
filme_amazon = amazon[fa]['type']
filme_amazon = filme_amazon.count()
sa = amazon['type'].isin(Show)
serie_amazon = amazon[sa]['type'].count()

a = {
    'Entretenimento' : ['Movies', 'TV Shows'],
     'valor' : [filme_amazon, serie_amazon]}
a = pd.DataFrame(a)
a
```

```python
sns.barplot(x = "Entretenimento", y = "valor", data = a)
```

Percebe-se que o padrão se mantém, sendo que a Amazon Prime investe mais no catálogo de filmes do que o de "séries de TV" proporcionalmente ao geral 

7.3 e 8.3 Netflix

```python
fn = netflix['type'].isin(Filme)
filme_netflix = netflix[fn]['type']
filme_netflix = filme_netflix.count()
sn = netflix['type'].isin(Show)
serie_netflix = netflix[sn]['type'].count()

n = {
    'Entretenimento' : ['Movies', 'TV Shows'],
     'valor' : [filme_netflix, serie_netflix]}
n = pd.DataFrame(n)
n
```
